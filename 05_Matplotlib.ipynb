```python
import matplotlib.pyplot as plt
df1=df.drop(["Id"],axis=1)  #ID den kurtulmak için yaptık
df1
```

```python
df1.plot()
plt.show()
```

```python
setosa=df[df.Species=="Iris-setosa"]
print(setosa.head())
print(setosa.tail())
print("---------------------------------------------------------")
versicolor=df[df.Species=="Iris-versicolor"]
print(versicolor.head())
print(versicolor.tail())
print("---------------------------------------------------------")
virginica=df[df.Species=="Iris-virginica"]
print(virginica.head())
print(virginica.tail())
```

```python
plt.plot(setosa.Id,setosa.PetalLengthCm,color="red",label="Setosa")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```

```python
plt.plot(versicolor.Id,versicolor.PetalLengthCm,color="green",label="Versicolor")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```

```python
plt.plot(virginica.Id,virginica.PetalLengthCm,color="blue",label="Virginica")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```

```python
plt.plot(setosa.Id,setosa.PetalLengthCm,color="red",label="Setosa")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()

plt.plot(versicolor.Id,versicolor.PetalLengthCm,color="green",label="Versicolor")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()

plt.plot(virginica.Id,virginica.PetalLengthCm,color="blue",label="Virginica")
plt.xlabel("Id")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```

```python
df1.plot(grid=True,alpha=0.5,linestyle=":")
plt.show()
```

```python
#--------------------SCATTER PLOT--------------------
```

```python
plt.scatter(setosa.PetalLengthCm,setosa.PetalWidthCm,color="red",label="Setosa")
plt.scatter(versicolor.PetalLengthCm,versicolor.PetalWidthCm,color="green",label="Versicolor")
plt.scatter(virginica.PetalLengthCm,virginica.PetalWidthCm,color="blue",label="Virginica")
plt.legend()
plt.xlabel("PetalLengthCm")
plt.xlabel("PetalWidthCm")
plt.title("Scatter Plot")
plt.show()
```

```python
#--------------------HISTOGRAM PLOT--------------------
```

```python
plt.hist(setosa.PetalLengthCm)
plt.xlabel("PetalLengthCm values")
plt.ylabel("Frequency")
plt.title("Histogram Plot")
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5,6,7])
a=["TR","USA","a","b","c","d","e"]
y=x*2+5
plt.bar(a,y)
plt.title("Bar Plot")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
setosa=df[df.Species=="Iris-setosa"]
versicolor=df[df.Species=="Iris-versicolor"]
virginica=df[df.Species=="Iris-virginica"]

plt.subplot(2,1,1)
plt.plot(setosa.Id,setosa.PetalLengthCm,color="red",label="Setosa")
plt.xlabel("ID")
plt.ylabel("Setosa-PetalLengthCm")

plt.subplot(2,1,2)
plt.plot(versicolor.Id,versicolor.PetalLengthCm,color="green",label="Versicolor")
plt.xlabel("ID")
plt.ylabel("Versicolor-PetalLengthCm")
plt.show()
```
