```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("random_forest_regression_dataset.csv",sep=";",header=None)
df
```

```python
x=df.iloc[:,0].values.reshape(-1,1)
y=df.iloc[:,1].values.reshape(-1,1)
print("----------------X Values----------------")
print(x)
print("----------------Y Values----------------")
print(y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x,y)
y_head=rf.predict(x)
```

```python
from sklearn.metrics import r2_score
print("r_score:",r2_score(y,y_head)) #Burada gerek y deerleri ve predict ettiimiz y deerleri ile r_square deerlerini
                                     #karlatrm olduk ve bu deer 1'e ne kadar yaknsa o kadar iyidir.
```
